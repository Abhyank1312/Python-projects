```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
import pygwalker as pg
```

```python
df=pd.read_csv("hotel_bookings.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop(["lead_time","adr","agent","company","distribution_channel","market_segment","stays_in_weekend_nights","stays_in_week_nights","total_of_special_requests","arrival_date_week_number","arrival_date_day_of_month"],axis=1)

df
```

```python
df["country"]=df["country"].fillna(method="ffill")
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.loc[(df["reserved_room_type"]==df["assigned_room_type"]),"room type"]="desired"
df.loc[(df["reserved_room_type"]!=df["assigned_room_type"]),"room type"]="Not-desired"
df
```

```python
df.loc[(df["adults"]==2) & ((df["children"] >0 )| (df["babies"]>0)),"guest type"]="family"
df.loc[(df["adults"]==2),"guest type"]="couple"
df.loc[(df["adults"]>2),"guest type"]="group"
df.loc[(df["adults"]==1),"guest type"]="single"
df
```

```python
sns.countplot(x ="hotel",data=df , hue="is_canceled")
plt.title("Hotel  cancellation hotel type", fontsize=20,pad=15 )
plt.show()
```

```python
sns.countplot(x="required_car_parking_spaces",data=df , hue="is_canceled")
plt.title("Hotel  cancellation based on parking space", fontsize=20,pad=15 )
plt.show()
```

```python
sns.countplot(x ="room type",data=df , hue="is_canceled")
plt.title("Hotel  cancellation based on desired room", fontsize=20,pad=15 )
plt.show()
```

```python
sns.countplot(x ="guest type",  data=df , hue="is_canceled")
plt.title("Hotel  cancellation based on guset type", fontsize=20,pad=15 )
plt.show()
```

```python
sns.countplot(x ="arrival_date_month",data=df , hue="is_canceled")
plt.xticks(rotation =30)
plt.title("Hotel  cancellation basis on month", fontsize=20,pad=15 )
plt.show()
```

```python
gb=df.groupby("arrival_date_year").agg({"is_canceled":"count"})
gb
```

```python
plt.pie(gb["is_canceled"],labels=gb.index,autopct="%1.2f")
plt.title("Hotel  cancellation during 3 years", fontsize=20,pad=15 )
plt.show()
```

```python
sns.countplot(x ="booking_changes",data=df , hue="is_canceled")
plt.title("Hotel  cancellation based on booking changes", fontsize=20,pad=15 )
plt.show()
```
