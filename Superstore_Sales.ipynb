# filling missing values in data set -:


```python
db.fillna(value=db["Product Base Margin"].mean(),inplace =True)
```

```python
db.isnull().sum()
```

```python
db["order Year"] = db["Order Date"].dt.year
```

# Analyzing data and graph creation

# order priority

```python
# THIS IS NUMERICAL ANALYSIS OF 'COUNT OF ORDER PRIORITY'
db['Order Priority'].value_counts()               
                                                    # during value count if you only use count() funcition it gives total of the column.
                                                    #if you use value_counts() it give count according to distribution.
```

```python
db["Order Priority"].unique()                 # unique () function give us the name of sub-column  from the column
```

```python
db["Order Priority"]= db["Order Priority"].replace(to_replace = "Critical" ,value = "Critical ")
```

```python
# THIS IS GRIPHICAL ANALYSIS OF 'COUNT OF Order Priority"
plt.figure(figsize = (5,3))
sns.countplot(x="Order Priority", data=db)
plt.title("Count of Order Priority")
plt.show()                                                           

```

# shipping mode

```python
db["Ship Mode"].value_counts()
```

```python
x=db["Ship Mode"].value_counts().index
y= db['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.title("count of ship mode")
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.show

```

# Ship Mode on basis of Product Categotry

```python
plt.figure(figsize=(5,3))
plt.title("count of Ship Mode basis on Product Category")
sns.countplot(x="Ship Mode", data= db, hue="Product Category")
plt.show()
```

# Customer Segment

```python
plt.figure(figsize=(6,3))
plt.title("count of customer segment")
sns.countplot(x="Customer Segment",data = db)
plt.show()
```

```python
db["Sales"].value_counts().index[:5]
```

# Mean of Sales 

```python
from numpy import mean

plt.figure(figsize=(5,3))
plt.title(" Average sales based on product category")
sns.barplot(x="Product Category",y="Sales", data= db,estimator=mean)
plt.show()
```

# Product Category basis on its sub category

```python
plt.figure(figsize=(10,6))
plt.title("count of product category basis on office supplies")
sns.countplot(x="Product Category" , data =db[db["Product Category"]=="Office Supplies"] , hue= "Product Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.title("count of product category basis on Furniture")
sns.countplot(x="Product Category" , data =db[db["Product Category"]=="Furniture"] , hue= "Product Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.title("count of product category basis on Technology")
sns.countplot(x="Product Category" , data =db[db["Product Category"]=="Technology"] , hue= "Product Sub-Category")
plt.show()
```

# Product Container 

```python
plt.figure(figsize=(8,5))
plt.title("count of product container based on Product Category")
sns.countplot(x="Product Container" , data = db,hue = "Product Category")
plt.show()
```

# Product Base Margin

```python
plt.figure(figsize=(5,3))
plt.title("Product base margin base on Product catrgory")
sns.barplot(x= "Product Category",y="Product Base Margin",data = db,estimator=sum)
plt.show()
```

# Order per year

```python
plt.figure(figsize=(5,3))
plt.title("count of order per year")
sns.countplot(x= "order Year",data = db)
plt.show()
```

# Profit 

```python
plt.figure(figsize=(5,3))
plt.title("Profit basis on product category")
sns.barplot(x="Product Category" , y= "Profit" , data = db, estimator=sum)
plt.show()
```
